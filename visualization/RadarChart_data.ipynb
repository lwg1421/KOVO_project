현대캐피탈 선수들만 포함된 데이터프레임 생성

```python
top_hdc = top_all.loc[top_all["구단"]=="현대캐피탈"]
top_hdc
```

```python
top_hdc.columns
```

```python
top_hdc["공격 성공률"] = top_hdc["공격 성공"]/top_hdc["공격 시도"]
top_hdc["공격 효율"] = (top_hdc["공격 성공"] - top_hdc["공격 상대 블락"] - top_hdc["공격 범실"])/top_hdc["공격 시도"]
top_hdc["서브 성공률"] = top_hdc["서브 성공"]/top_hdc["서브 시도"]
top_hdc["세트 성공률"] = top_hdc["세트 성공"]/top_hdc["세트 시도"]
top_hdc["블로킹 성공률"] = top_hdc["블로킹 성공"]/top_hdc["블로킹 시도"]
top_hdc["리시브 효율"] = (top_hdc["리시브 정확"]-top_hdc["리시브 실패"])/top_hdc["리시브 시도"]
top_hdc["디그 성공률"] = top_hdc["디그 성공"]/top_hdc["디그 시도"]
top_hdc.columns
```

```python
top_hdc.head()
```

```python
top_hdc_radar = top_hdc[["선수명",'구단','경기수', '세트수', '득점','공격 성공률', '서브 성공률', '세트 성공률', '블로킹 성공률', 
'공격 효율', '리시브 효율', '디그 성공률']]
top_hdc_radar.reset_index(drop = True,inplace = True)
top_hdc_radar
```

```python
top_hdc_radar["공격 효율"].loc[top_hdc_radar["공격 효율"]<0] = 0
top_hdc_radar["리시브 효율"].loc[top_hdc_radar["리시브 효율"]<0] = 0
```

```python
top_hdc_radar_na = top_hdc_radar.fillna(0)
top_hdc_radar_na.rename(columns = {"선수명":"player_name"},inplace = True)
```

```python
top_hdc_radar_na
```

```python
player_data = pd.read_excel("C:/kovo/KOVO_player_data.xlsx")
player_data.head()
```

```python
# 선수정보/ 선수 기록 합친 데이터프레임 생성

player_data_merge = pd.merge(player_data, top_hdc_radar_na, how = "right", on="player_name")
#player_data_merge = player_data_merge.drop(columns = ["Unnamed: 0"])
```
