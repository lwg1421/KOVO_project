### 선수 개인 시즌별 기록 크롤러 제작

```python
import pandas as pd
import requests
import re
import warnings
from bs4 import BeautifulSoup as bs

# 불필요한 경고 출력을 방지합니다.
warnings.filterwarnings('ignore')
```

```python
# link_list에 각 선수별 링크 저장
link_list_1 = ['https://www.kovo.co.kr/team/21132_player_view.asp?p_code=1000059&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=0001178&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=0001377&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=0001365&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=0001462&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=0001967&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=1000138&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=0000853&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=0002255&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=0001873&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=0000874&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=0001956&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=0002256&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=1000139&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=0001291&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=0002233&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=0002104&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=1000060&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=0000027&t_code=1005&s_part=1', 'https://www.kovo.co.kr/team/21132_player_view.asp?p_code=1000061&t_code=1005&s_part=1']
```

```python
# 크롤러 제작
player_name = []
player_number = []
player_position = []
game = []
season = []
team = []
game_count = []
set_count = []
score = []
attack_succes_percent = []
bloocking_avg = []
serve_avg = []
set_avg = []
reveive_eff = []
dig_avg = []
penalty = []
mistake = []

for player_link in link_list_1 :
    url = player_link
    player_html = requests.get(url, verify = False).text
    player_soup = bs(player_html,"lxml")
    
    name = player_soup.find("p",{"class":"name"}).get_text()
    no = player_soup.find("p",{"class":"no"}).get_text()
    player_num = no.split("\xa0")[0]
    player_pos = no.split("\xa0")[1]

    player_record_table = player_soup.find("table",{"class":"lst_board"})
    player_record_data = player_record_table.select("tbody > tr > td")

    player_record_data_real = []

    for dt in player_record_data :
        player_record_data_real.append(str(dt.get_text()))

        for i in range(len(player_record_data_real)//14):
            player_name.append(name)
            player_number.append(player_num)
            player_position.append(player_pos)
            mistake.append(player_record_data_real.pop())
            penalty.append(player_record_data_real.pop())
            dig_avg.append(player_record_data_real.pop())
            reveive_eff.append(player_record_data_real.pop())
            set_avg.append(player_record_data_real.pop())
            serve_avg.append(player_record_data_real.pop())
            bloocking_avg.append(player_record_data_real.pop())
            attack_succes_percent.append(player_record_data_real.pop())
            score.append(player_record_data_real.pop())
            set_count.append(player_record_data_real.pop())
            game_count.append(player_record_data_real.pop())
            team.append(player_record_data_real.pop())
            season.append(player_record_data_real.pop())
            game.append(player_record_data_real.pop())
            n = len(player_record_data_real)
```

```python
# 데이터프레임 생성
player_record = pd.DataFrame({"player_name":player_name,"player_position":player_position,"player_number":player_number,"game":game,"season":season,"game_count":game_count,"set_count":set_count,"score":score,"attack_succes_percent":attack_succes_percent,
"bloocking_avg":bloocking_avg,"serve_avg":serve_avg,"set_avg":set_avg,"reveive_eff":reveive_eff,"dig_avg":dig_avg,"penalty":penalty,"mistake":mistake})
```
